```python
def readhtml(roomurl):
    rand=random.randint(0,len(user_agent)-1)
    header={'user-agent':user_agent[rand]}
    req = urllib2.Request(roomurl,headers=header)
    response=urllib2.urlopen(req)
    html=response.read()
    
    soup = BeautifulSoup(html, 'lxml')
        
    return soup
def get_title(soup):
    T=soup.find('title').get_text()
    T=T.split('-')[0]
    return T
def get_price_url(allid):
    priurl="https://www.airbnb.cn/api/v2/pdp_listing_booking_details?force_boost_unc_priority_message_type=&guests=1&listing_id={}&non_refundable_policy_selected=0&show_smart_promotion=0&_format=for_web_dateless&_interaction_type=pageload&_intents=p3_book_it&_parent_request_uuid=19debf2b-6621-495f-9857-6147f2349183&_p3_impression_id=p3_1532072882_8XZ%2FiBrNFofLq6Jd&number_of_adults=1&number_of_children=0&number_of_infants=0&key=d306zoyjsyarp7ifhu67rjxn52tv0t20&currency=CNY&locale=zh".format(allid) 
    return priurl
def get_price(priceurl):
    url=priceurl
    response = urllib2.urlopen(url)
    html=response.read()
    data2=json.loads(html)
    price1=data2['pdp_listing_booking_details'][-1]['p3_display_rate']['amount']
    return price1
def get_review(soup,eachroomurl):
    try:
        maxpage=soup.find_all(class_="_1bdke5s")[-1].get_text()
    except:
        maxpage=1
    RE=[] 
    for page in range(int(maxpage)):
        eachreviewurl='https://www.airbnb.cn/api/v2/reviews?key=d306zoyjsyarp7ifhu67rjxn52tv0t20&currency=CNY&locale=zh&listing_id={}&role=guest&_format=for_p3&_limit=7&_offset={}&_order=language_country'.format(eachid,page*7)    
        print eachreviewurl
        rand=random.randint(0,len(user_agent)-1)
        header={'user-agent':user_agent[rand]}
        req = urllib2.Request(eachreviewurl,headers=header)
        response=urllib2.urlopen(req)
        html=response.read()
        data2=json.loads(html)
    #    print data2
        count=len(data2["reviews"])
        for c in range(count):
            review=[]
        #    print c
            Iid=data2["reviews"][c]["id"]
        #    print Iid
            review.append(Iid)
            review1=data2["reviews"][c]["comments"]
         #   print review1
            review.append(review1)
            RE.append(review)
        time.sleep(0.5)
    return RE
```

```python
#价钱最高的前10
import pprint
query = """SELECT id,name,price 
            FROM Airbnb1
            
            GROUP BY price
            ORDER BY price
            DESC
            LIMIT 10;
            """
data = list()
for row in conn1.execute(query):
    data.append(row)
pprint.pprint(data)
```

```python
#单身公寓下价钱最高的前10
query = """SELECT id,name,price 
            FROM Airbnb1
            WHERE bedroom='单间公寓'
            GROUP BY price
            ORDER BY price
            DESC
            LIMIT 10;
            """
data = list()
for row in conn1.execute(query):
    data.append(row)
pprint.pprint(data)
```

```python
#查询有5张床或者能容纳5人的房间，从价格高往下排
query = """SELECT id,name,bed, maxmember,price
            FROM Airbnb1
            WHERE (bed='5' or maxmember='5' )
            GROUP BY price
            ORDER BY price
            DESC
            LIMIT 10;
            """
data = list()
for row in conn1.execute(query):
    data.append(row)
pprint.pprint(data)
```
